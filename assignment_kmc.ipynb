**Q1.** This question is a case study for $k$ means clustering.

1. Load the `airbnb_hw.csv` data. Clean `Price` along with `Beds`, `Number of Reviews`, and `Review Scores Rating`.
2. Maxmin normalize the data and remove any `nan`'s (`KMeans` from `sklearn` doesn't accept `nan` input).
3. Use `sklearn`'s `KMeans` module to cluster the data by `Beds`, `Number of Reviews`, and `Review Scores Rating` for `k=6`.
4. Use `seaborn`'s `.pairplot()` to make a grid of scatterplots that show how the clustering is carried out in multiple dimensions.
5. Use `.groupby` and `.describe` to compute the average price for each cluster. Which clusters have the highest rental prices?
6. Use a scree plot to pick the number of clusters and repeat steps 4 and 5.

````python
import pandas as pd

df = pd.read_csv('kmc/data/airbnb_hw.csv')
X = df.loc[:,['Price','Beds','Number Of Reviews','Review Scores Rating']]

print(X.shape)
X.describe()
````

````python
X['Price'].value_counts()
X['Price'] = X['Price'].str.replace(',','')
X['Price'] = pd.to_numeric(X['Price'],errors='coerce')
X.describe()
````

````python
X['Beds'] = X['Beds'].fillna(1)
X.describe()
````

````python
pd.crosstab(df['Review Scores Rating'].isnull(), df['Number Of Reviews']>0)
````

````python
X = X.dropna()
X.describe()
````

````python
def maxmin(x):
    u = (x-min(x))/(max(x)-min(x))
    return u

Z = X.drop('Price',axis=1)
Z = Z.apply(maxmin)
````

````python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

model = KMeans(n_clusters=6, max_iter=300, n_init = 10, random_state=0)
model = model.fit(Z)
Z['cluster'] = model.labels_
Z.describe()
````

````python
import seaborn as sns
sns.pairplot(data=Z,hue='cluster')
````

````python
X['cluster'] = model.labels_
X.loc[:,['Price','cluster'] ].groupby('cluster').describe()
````

Cluster 1 seems to have the highest rental prices based on the mean value of 293.46, which is much higher than the other clusters which have a mean around 130. Looking at the 25/50/75 percentiles, we see that outliers don't affect this too much as cluster 1 has the highest values across the distribution.



````python
import numpy as np

k_bar = 15
k_grid = np.arange(1,k_bar+1)
SSE = np.zeros(k_bar)
for k in range(k_bar):
    model = KMeans(n_clusters=k+1, max_iter=300, n_init = 10, random_state=0)
    model = model.fit(Z)
    SSE[k] = model.inertia_

sns.lineplot(x=k_grid, y=SSE).set_title('Scree Plot')
````

````python
model = KMeans(n_clusters=3, max_iter=300, n_init = 10, random_state=0)
model = model.fit(Z)
Z['cluster'] = model.labels_

sns.pairplot(data=Z,hue='cluster')
X['cluster'] = model.labels_
X.loc[:,['Price','cluster'] ].groupby('cluster').describe()
````

Cluster 2 has the highest reviews based on the mean value of 194.83 and also when looking at the distribution at the 25/50/75 percentiles.

1. Run the code below, which creates four datasets: `df0_125`, `df0_25`, `df0_5`, `df1_0`, and `df2_0`. Each data set is created by increasing the amount of `noise` (standard deviation) around the cluster centers, from `0.125` to `0.25` to `0.5` to `1.0` to `2.0`.

2. Make scatterplots of the $(X1,X2)$ points by group for each of the datasets. As the `noise` goes up from 0.125 to 2.0, what happens to the visual distinctness of the clusters?
3. Create a scree plot for each of the datasets. Describe how the level of `noise` affects the scree plot (particularly the presence of a clear "elbow") and your ability to definitively select a $k$.
4. Explain the intuition of the elbow, using this numerical simulation as an example.

````python
import numpy as np
import pandas as pd

def createData(noise,N=50):
    np.random.seed(100) # Set the seed for replicability
    # Generate (x1,x2,g) triples:
    X1 = np.array([np.random.normal(1,noise,N),np.random.normal(1,noise,N)])
    X2 = np.array([np.random.normal(3,noise,N),np.random.normal(2,noise,N)])
    X3 = np.array([np.random.normal(5,noise,N),np.random.normal(3,noise,N)])
    # Concatenate into one data frame
    gdf1 = pd.DataFrame({'x1':X1[0,:],'x2':X1[1,:],'group':'a'})
    gdf2 = pd.DataFrame({'x1':X2[0,:],'x2':X2[1,:],'group':'b'})
    gdf3 = pd.DataFrame({'x1':X3[0,:],'x2':X3[1,:],'group':'c'})
    df = pd.concat([gdf1,gdf2,gdf3],axis=0)
    return df

df0_125 = createData(0.125)
df0_25 = createData(0.25)
df0_5 = createData(0.5)
df1_0 = createData(1.0)
df2_0 = createData(2.0)
````

````python
sns.scatterplot(data = df2_0, x = 'x1',y='x2',hue='group',style='group')
````

As the noise goes up from 0.125 to 2.0, the clusters appear to spread out and become more intersected with each other. It gets harder to distinguish where each cluster begins and ends as the noise increases.

````python
def scree(data):
    X = data.loc[ : , ['x1','x2'] ]
    X = X.apply(maxmin)
    k_bar = 15
    k_grid = np.arange(1,k_bar+1)
    SSE = np.zeros(k_bar)
    for k in range(k_bar):
        model = KMeans(n_clusters=k+1, max_iter=300, n_init = 10, random_state=0)
        model = model.fit(X)
        SSE[k] = model.inertia_
    scree_plot, axes = plt.subplots()
    sns.lineplot(x=k_grid, y=SSE).set_title('Scree Plot')
    axes.set_ylim(0, 35)

scree( data = df0_125 )
````

````python
scree( data = df0_25 )
````

````python
scree( data = df0_5 )
````

````python
scree( data = df1_0 )
````

````python
scree( data = df2_0 )
````

As the level of noise increases, it becomes harder to clearly identify where the elbow is in the plot. For the first three plots we can see that an elbow is present at k=3, but for the last two it becomes harder to tell.

When data groups are clearly separated and distinct, as seen in the dataset with the lowest noise level (0.125), the addition of another cluster (increasing k) results in a substantial decrease in the sum of squared errors (SSE). This decrease indicates that each new cluster meaningfully captures a distinct group of data points, thereby improving the overall clustering model's fit to the data. However, as the noise increases, which causes the data groups to spread out and overlap, the visual distinctness of these groups diminishes. This overlap makes it harder to discern how many groups are present and to which group each data point is in. In such scenarios, the benefit of adding more clusters beyond the true number of data groups (in this case, three) decreases, leading to a smoother transition in the SSE as k increases. This causes the scree plot's elbow, which signals the optimal number of clusters, to become less definite and noticeable. When the data is visualized like a "cloud" without clear distinctions between groups, identifying a distinct elbow becomes difficult, if not impossible.
